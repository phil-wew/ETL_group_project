```python
import pandas as pd
import numpy as np

census= pd.read_csv(r'census_city.csv')

census_pop= pd.DataFrame(census)

census_pop
```

```python
census_pop.loc[:,'City'] = census_pop.City.map(lambda x : x.replace('city', ''))
```

```python
census_pop

census_pop.to_csv(r'census_pop.csv')
```
